```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
```

```python
plt.rcParams['font.family']='gulim'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 업종별 만족도와 관심도에 대한 분석

# 선정된 각 자치구별 평점 상위 5개 업종에 대한 평점 평균과 조회수를 비교하여 실제 이용고객의 만족도와 (잠재)고객의 관심도를 알아보고자
# 하였으며,보조 수치로 방문고객의 리뷰수를 참고하여 산점도로 나타냄
```

```python
food = ['이자카야/오뎅/꼬치', '이탈리안', '카페/디저트', '칵테일/와인', '프랑스음식']

score = [4.5,4.447368421,4.446666667,4.45,4.45]
review = [845,1482,1547,164,496]

view = [2650077,2166410,1442472,231328,1211589]

size = np.array(view)*0.005
colors = ['lightblue','pink','yellow','lightgreen','lightgray']

plt.xlabel('Review', labelpad=20)
plt.ylabel('Score', labelpad=20)
plt.title('강남구', loc='center', pad=30)

for x, y, name in zip(review, score, view):
    plt.text(x, y, name)

for i in range(5):
    plt.scatter(review, score, label = food[i], s=size, c=colors)
    
plt.legend(loc=(1.03,0.64), markerscale = 0.05)
plt.show()
```

```python
# 강남구
## '이자카야/오뎅/꼬치' 업종의 만족도와 관심도가 모두 높았으며, 주류가 수반되는 업종은 자치구나 주종과 무관하게 리뷰가 
## 상대적으로 적게 등록되는 것으로 보임
## '카페/디저트'나 '이탈리안' 은 높은 리뷰와 조회수로 보아 적극적인 고객이 많고 고객의 관심도가 큰 업종으로 보이므로 여타 다른 분석을
## 추가하여 업종크기(업장 수)를 늘리거나 서비스개선을 통해 고객의 만족도를 높인다면 신규 고객의 유입을 기대할 수 있을 것으로 판단됨
## 특히 '이탈리안'은 프랑스음식과 유사한 업종으로, 평점에 있어 프랑스음식보다 근소하게 낮은 점수를 기록했으나 리뷰나 조회수에 있어
## 월등한 관심도를 보이므로 다각도로 평점을 높이는 노력을 한다면 프랑스음식에 대한 고객 수요를 가져올 수 있을 것으로 판단됨
```

```python
food = ['까스 요리','라멘/소바/우동','이자카야/오뎅/꼬치','이탈리안','일식']

score = [4.425,4.404545455,4.408333333,4.447058824,4.44]
review = [717,2481,533,1231,231]

view = [530031,1956860,1208071,1821966,281854]

size = np.array(view)*0.005
colors = ['lightblue','pink','yellow','lightgreen','lightgray']

plt.xlabel('Review', labelpad=20)
plt.ylabel('Score', labelpad=20)
plt.title('마포구', loc='center', pad=30)

for x, y, name in zip(review, score, view):
    plt.text(x, y, name)

for i in range(5):
    plt.scatter(review, score, label = food[i], s=size, c=colors)
    
plt.legend(loc=(1.03,0.64), markerscale = 0.05)
plt.show()
```

```python
# 마포구
## 일식과 이탈리안 업종의 평점은 근소한 차이를 보인 반면, 리뷰와 조회수는 이탈리안이 월등히 높음. 이탈리안업종에 대한 고객들의 만족도와 
## 관심도가 모두 높으므로 서비스품질유지 차원에서의 노력이 필요해 보임.
## 고객의 관심도가 가장 높은 업종은 라멘/소바/우동으로 높은 리뷰수까지 고려할 때 평점향상을 위한 개선이 필요함
```

```python
food = ['국수/면요리', '기타양식','프랑스음식', '해산물요리', '회/스시']

score = [4.355555556,4.26,4.36,4.3,4.266666667]
review = [660,214,130,66,313]

view = [577701,	225754,256750,90163,795217]

size = np.array(view)*0.005
colors = ['lightblue','pink','yellow','lightgreen','lightgray']

plt.xlabel('Review', labelpad=20)
plt.ylabel('Score', labelpad=20)
plt.title('서초구', loc='center', pad=30)

for x, y, name in zip(review, score, view):
    plt.text(x, y, name)

for i in range(5):
    plt.scatter(review, score, label = food[i], s=size, c=colors)
    
plt.legend(loc=(1.06,0.64), markerscale = 0.05)
plt.show()
```

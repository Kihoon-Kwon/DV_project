```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df_Gangnam = pd.read_csv("C:/Users/user/Desktop/DV_project/csv_files/Gangnam.csv",index_col="이름",encoding='cp949')
df_Mapo = pd.read_csv("C:/Users/user/Desktop/DV_project/csv_files/Mapo.csv",index_col="이름",encoding='cp949')
df_Seocho = pd.read_csv("C:/Users/user/Desktop/DV_project/csv_files/Seocho.csv",index_col="이름",encoding='cp949')

plt.rc("font", family="Malgun Gothic")
```

```python
Gangnam = df_Gangnam.sort_values("카테고리")
Mapo = df_Mapo.sort_values("카테고리")
Seocho = df_Seocho.sort_values("카테고리")

Gangnam['조회 수'] = Gangnam['조회 수'].str.replace(',','').astype('int64')
Mapo['조회 수'] = Mapo['조회 수'].str.replace(',','').astype('int64')
Seocho['조회 수'] = Seocho['조회 수'].str.replace(',','').astype('int64')
```

```python
Seocho_foodlist = list(df_Seocho["카테고리"])
Seocho_foodlist = set(Seocho_foodlist)
Seocho_categories = list(Seocho_foodlist) 
Seocho_categories.sort() # 서초구 음식점 리스트(오름차순)

Mapo_foodlist = list(df_Mapo["카테고리"])
Mapo_foodlist = set(Mapo_foodlist)
Mapo_categories = list(Mapo_foodlist) 
Mapo_categories.sort() # 마포구 음식점 리스트(오름차순)

Gangnam_foodlist = list(df_Gangnam["카테고리"])
Gangnam_foodlist = set(Gangnam_foodlist)
Gangnam_categories = list(Gangnam_foodlist) 
Gangnam_categories.sort() # 강남구 음식점 리스트(오름차순)
```

```python
Seocho_scores = [] # 서초구 음식점 종류 별 평점 평균
Seocho_views = [] # 서초구 음식점 종류 별 조회수 합
Seocho_reviews = [] # 서초구 음식점 종류 별 리뷰수 합

for s in Seocho_categories:
    df=Seocho[Seocho['카테고리']==s].copy() # 각 음식 종류별로 df에 저장
    score = sum(list(df["평점"]))/len(list(df["평점"]))
    review = sum(list(df["리뷰 수"]))
    view = sum(list(df['조회 수']))
    
    Seocho_scores.append(round(score,2))
    Seocho_views.append(view)
    Seocho_reviews.append(review)
```

```python
Mapo_scores = [] # 마포구 음식점 종류 별 평점 평균
Mapo_views = [] # 마포구 음식점 종류 별 조회수 합
Mapo_reviews = [] # 마포구 음식점 종류 별 리뷰수 합

for m in Mapo_categories:
    df=Mapo[Mapo['카테고리']==m].copy() # 각 음식 종류별로 df에 저장
    score = sum(list(df["평점"]))/len(list(df["평점"]))
    review = sum(list(df["리뷰 수"]))
    view = sum(list(df['조회 수']))
    
    Mapo_scores.append(round(score,2))
    Mapo_views.append(view)
    Mapo_reviews.append(review)
```

```python
Gangnam_scores = [] # 강남구 음식점 종류 별 평점 평균
Gangnam_views = [] # 강남구 음식점 종류 별 조회수 합
Gangnam_reviews = [] # 강남구 음식점 종류 별 리뷰수 합

for g in Gangnam_categories:
    df=Gangnam[Gangnam['카테고리']==g].copy() # 각 음식 종류별로 df에 저장
    score = sum(list(df["평점"]))/len(list(df["평점"]))
    review = sum(list(df["리뷰 수"]))
    view = sum(list(df['조회 수']))
    
    Gangnam_scores.append(round(score,2))
    Gangnam_views.append(view)
    Gangnam_reviews.append(review)
```

```python
# 강남구

color_total = []

for i in range(50):
    x = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])]
    color_total.append(x[0])
    
color = np.random.choice(color_total, len(Gangnam_categories), replace=False)
    
size = np.array(Gangnam_reviews)*0.8

fig = plt.figure(figsize=(10,6))

plt.scatter(
        x = Gangnam_reviews,
        y = Gangnam_scores,
        s=size,
        c=color,
        cmap='Accent',
        alpha=0.8,
        edgecolor='gray',
        linewidth=1)

plt.xlabel('리뷰')
plt.ylabel('평점')
plt.title('강남구의 리뷰 수와 평점의 상관관계')
plt.grid(True, which='major', linestyle='--', linewidth=0.5)

plt.savefig('Gangnam.png',dpi =100)
plt.show()
```
